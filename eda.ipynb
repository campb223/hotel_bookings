```python
# Calculate the average lead time
avg_lead_time = hotel_bookings['lead_time'].mean()

# Print the average lead time with 2 decimal places
print('The average lead time is: {:.2f}'.format(avg_lead_time))
```

```python
# Calculate the percentage of cancelled bookings
cancelled_bookings = hotel_bookings['is_canceled'].sum()
total_bookings = len(hotel_bookings)
cancel_percentage = cancelled_bookings / total_bookings * 100

# Print the cancellation percentage
print('The percentage of bookings that were cancelled is: {:.2f}%'.format(cancel_percentage))
```

```python
# Count the number of repeat guests
num_repeat_guests = len(hotel_bookings[hotel_bookings['is_repeated_guest'] == 1])

# Print the number of repeat guests
print('The number of repeat guests is:', num_repeat_guests)

# Calculate the percentage of repeat guests
repeat_percentage = num_repeat_guests / total_bookings * 100

# Print the percentage of repeat guests
print('The percentage of bookings made by repeat guests is: {:.2f}%'.format(repeat_percentage))
```

```python
# Calculate the average length of stay
avg_length_of_stay = hotel_bookings[['stays_in_weekend_nights', 'stays_in_week_nights']].sum(axis=1).mean()

# Print the average length of stay
print('The average length of stay is: {:.2f} nights'.format(avg_length_of_stay))
```

```python
# Group the data by resort type and sum the relevant columns
resort_bookings = hotel_bookings.groupby('hotel').sum(numeric_only=True)[['adults', 'children', 'babies']]

# Print the sum of different resort types
print('Total number of guests by resort type:')
print(resort_bookings)
```

```python
# Group the data by deposit type and sum the relevant columns
deposit_bookings = hotel_bookings.groupby('deposit_type').sum(numeric_only=True)[['adults', 'children', 'babies']]

# Print the sum of different deposit types
print('Total number of guests by deposit type:')
print(deposit_bookings)
```

```python
# Calculate the total number of stays
total_stays = len(hotel_bookings)

# Calculate the sum of adults, children, and babies
sum_adults = hotel_bookings['adults'].sum()
sum_children = hotel_bookings['children'].sum()
sum_babies = hotel_bookings['babies'].sum()

# Calculate the average number of adults, children, and babies per stay
avg_adults = sum_adults / total_stays
avg_children = sum_children / total_stays
avg_babies = sum_babies / total_stays

# Print the results
print('The average number of adults per stay is: {:.2f}'.format(avg_adults))
print('The average number of children per stay is: {:.2f}'.format(avg_children))
print('The average number of babies per stay is: {:.2f}'.format(avg_babies))
```

```python
# Filter the data for stays by adults only
adult_bookings = hotel_bookings[hotel_bookings['adults'] > 0]

# Calculate the average stays in weekend nights and week nights separately for adults
avg_weekend_stays = adult_bookings['stays_in_weekend_nights'].mean()
avg_weekday_stays = adult_bookings['stays_in_week_nights'].mean()

# Print the results
print('The average number of weekend nights stayed by adults is: {:.2f}'.format(avg_weekend_stays))
print('The average number of weekday nights stayed by adults is: {:.2f}'.format(avg_weekday_stays))
```

```python
# Count the frequency of each month
month_counts = hotel_bookings['arrival_date_month'].value_counts()

# Define the order of the months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Sort the months in chronological order
month_counts = month_counts.loc[month_order]

# Plot the bar graph
month_counts.plot(kind='bar')

# Add labels and title to the plot
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.title('Distribution of Bookings by Month')

# Show the plot
plt.show()
```

```python
# Plot the histogram
plt.hist(hotel_bookings['arrival_date_day_of_month'], bins=31)

# Add labels and title to the plot
plt.xlabel('Day of Month')
plt.ylabel('Number of Bookings')
plt.title('Distribution of Bookings by Day of Month')

# Show the plot
plt.show()
```

```python
# Plot the histogram
plt.hist(hotel_bookings['arrival_date_year'], bins=3)

# Add labels and title to the plot
plt.xlabel('Year')
plt.ylabel('Number of Bookings')
plt.title('Distribution of Bookings by Year')

# Show the plot
plt.show()
```
